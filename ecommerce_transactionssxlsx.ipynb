#### RFM Analysis

```python
reference_date = df['Transaction_Date'].max() + timedelta(days=2)
print(reference_date)
```

```python
rfm = df.groupby('User_Name').agg({
    'Transaction_Date': lambda x: (reference_date - x.max()).days,
    'Transaction_ID': 'count',
    'Purchase_Amount': sum
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head(5)
```

```python
quantile = rfm[['Recency', 'Frequency', 'Monetary']].quantile(q=[0.2, 0.4, 0.6, 0.8])

def r_score(x, q):
    if x <= q[0.2]:
        return 5
    elif x <= q[0.4]:
        return 4
    elif x <= q[0.6]:
        return 3
    elif x <= q[0.8]:
        return 2
    else:
        return 1

def fm_score(x, q):
    if x <= q[0.2]:
        return 1
    elif x <= q[0.4]:
        return 2
    elif x <= q[0.6]:
        return 3
    elif x <= q[0.8]:
        return 4
    else:
        return 5


rfm['R_Score'] = rfm['Recency'].apply(r_score, args=(quantile['Recency'],))
rfm['F_Score'] = rfm['Frequency'].apply(fm_score, args=(quantile['Frequency'],))
rfm['M_Score'] = rfm['Monetary'].apply(fm_score, args=(quantile['Monetary'],))
```

```python
rfm.head()
```

#### RFM SCORE

```python
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
```

#### Customer Segmentation

```python
rfm['RFM_Sum_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

def segment(x):
    if x >= 11:
        return 'Top Customer'
    elif x >= 6:
        return 'Potential'
    else:
        return 'Low Value'

rfm['Customer_Segmentation'] = rfm['RFM_Sum_Score'].apply(segment)
```

```python
rfm[['RFM_Score', 'Customer_Segmentation']].sample(10)
```

```python
segment_counts = rfm['Customer_Segmentation'].value_counts()

segment_counts.plot(
    kind='bar',
    color='cornflowerblue',
    figsize=(6, 4)
)
plt.title('Customer Segmentation Breakdown')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

#### Age Segmentation:  Buying Pattern Across Age Group

```python
bins = [17, 25, 45, 65, 71]
labels = ['1824', '2544', '4564', '6570']

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels,
    right=True,
)

revenue_age = df.pivot_table(index='Age_Group', values='Purchase_Amount', aggfunc='sum')
popular_age = df.pivot_table(index='Age_Group', values='Transaction_ID', aggfunc='count')
```

```python
#revenue_age = df.groupby('Age_Group')['Purchase_Amount'].sum()

revenue_age.plot(kind='bar', color='cornflowerblue', title='Revenue by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Revenue')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
